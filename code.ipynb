```python
import matplotlib.pyplot as plt
from sklearn.metrics import roc_curve, auc

# Assuming y_pred_prob contains probability scores of the positive class
y_pred_prob = dt_classifier.predict_proba(X_test)[:, 1]

# Compute ROC curve and ROC area
fpr, tpr, _ = roc_curve(Y_test, y_pred_prob)
roc_auc = auc(fpr, tpr)

# Plotting the ROC curve
plt.figure(figsize=(8, 6))
plt.plot(fpr, tpr, color='darkorange',
         lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic (ROC) Curve for Decision tree')
plt.legend(loc="lower right")
plt.show()
```

***Area Under the Curve (AUC)***:

AUC value is noted as 0.66.
The AUC measures the entire two-dimensional area underneath the entire ROC curve from (0,0) to (1,1).
It provides an aggregate measure of performance across all possible classification thresholds. The higher the AUC, the better the model is at distinguishing between the classes.
AUC values range from 0 to 1. A model with 100% correct predictions has an AUC of 1

# **Hybrid Algorithms(Decisiontree+Guassian)**

```python
from sklearn.ensemble import VotingClassifier

# Initialize classifiers
dt_classifier = DecisionTreeClassifier(random_state=42)
nb_classifier = GaussianNB()

# Voting classifier for combining Decision Tree and Naive Bayes
hybrid_classifier = VotingClassifier(estimators=[
    ('decision_tree', dt_classifier),
    ('naive_bayes', nb_classifier)
], voting='soft')

# Train the classifier
hybrid_classifier.fit(X_train, Y_train)

# Predict on the test set
y_pred_hybrid = hybrid_classifier.predict(X_test)

# Calculate accuracy
accuracy_hybrid = accuracy_score(Y_test, y_pred_hybrid)
print(f'Hybrid DT-NB Accuracy: {accuracy_hybrid:.2f}')
```

```python
# Confusion matrix
conf_matrix_hybrid = confusion_matrix(Y_test, y_pred_hybrid)
print('Hybrid DT-NB Confusion Matrix:')
print(conf_matrix_hybrid)
```

```python
conf_matrix_hybrid = confusion_matrix(Y_test, y_pred_hybrid)
sns.heatmap(conf_matrix_hybrid, annot=True, fmt='d', cmap='Blues')
plt.title('Hybrid DT-NB Confusion Matrix')
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.show()
```

True Positives (TP): The top left cell (1926) represents the count of positive cases correctly identified by the model.

```python
# Assuming the hybrid model can produce probability estimates
y_pred_prob_hybrid = hybrid_classifier.predict_proba(X_test)[:, 1]
fpr_hybrid, tpr_hybrid, _ = roc_curve(Y_test, y_pred_prob_hybrid)
roc_auc_hybrid = auc(fpr_hybrid, tpr_hybrid)

plt.figure(figsize=(8, 6))
plt.plot(fpr_hybrid, tpr_hybrid, color='darkorange', lw=2, label=f'ROC curve (area = {roc_auc_hybrid:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve for Hybrid DT-NB')
plt.legend(loc="lower right")
plt.show()
```

True Positive Rate (also known as recall or sensitivity) measures the proportion of actual positives that are correctly identified as such.

False Positive Rate measures the proportion of actual negatives that are incorrectly classified as positives.

The orange line represents the ROC curve for the Hybrid DT-NB model, which shows the trade-off between the True Positive Rate and False Positive Rate at different thresholds. The area under the curve (AUC) is a measure of the model's ability to distinguish between the classes and is indicated as 0.89. This value ranges from 0 to 1, where a value closer to 1 indicates a better model that can accurately distinguish between the positive and negative classes.

The blue dashed line represents a no-skill classifier that acts as a baseline, where the true positive rate equals the false positive rate. The ROC curve lying above this line suggests that the classifier provides a better-than-random prediction for the classes. The closer the ROC curve is to the upper left corner, the higher the overall accuracy of the test in correctly classifying the cases, reflecting a desirable outcome in most diagnostic settings.

# ***Hybrid Algorithm of XGBoost and ANN***

```python
from xgboost import XGBClassifier
from keras.models import Sequential
from keras.layers import Dense
from sklearn.metrics import accuracy_score

# Define and train XGBoost model
xgb_model = XGBClassifier()
xgb_model.fit(X_train, Y_train)

# Define and train ANN model
ann_model = Sequential([
    Dense(128, activation='relu', input_dim=X_train.shape[1]),
    Dense(64, activation='relu'),
    Dense(1, activation='sigmoid')
])
ann_model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
ann_model.fit(X_train, Y_train, epochs=10, batch_size=32, verbose=0)  # Adjust verbose as needed

# Evaluate XGBoost model
xgb_y_pred = xgb_model.predict(X_test)
xgb_accuracy = accuracy_score(Y_test, xgb_y_pred)
print(f'XGBoost Model Accuracy: {xgb_accuracy:.2f}')

# Evaluate ANN model
ann_y_pred_prob = ann_model.predict(X_test)  # Predict probabilities
ann_y_pred = (ann_y_pred_prob > 0.5).astype(int).flatten()  # Convert probabilities to binary class labels
ann_accuracy = accuracy_score(Y_test, ann_y_pred)
print(f'ANN Model Accuracy: {ann_accuracy:.2f}')

# Combine predictions (example ensemble)
combined_y_pred = (xgb_y_pred + ann_y_pred) // 2  # Simple averaging for binary classification
combined_accuracy = accuracy_score(Y_test, combined_y_pred)
print(f'Combined Model Accuracy: {combined_accuracy:.2f}')
```

# ***CNN(convolutional neural network) algorithm to the dataset***

```python
from keras.models import Sequential
from keras.layers import Conv1D, MaxPooling1D, Flatten, Dense
from keras.optimizers import Adam
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
import numpy as np

# Assuming X and Y are already defined
X = np.expand_dims(X, axis=2)

# Split the data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)

# Define the 1D CNN model architecture
cnn_model = Sequential([
    Conv1D(32, 3, activation='relu', input_shape=(X_train.shape[1], X_train.shape[2])),
    MaxPooling1D(2),
    Conv1D(64, 3, activation='relu'),
    MaxPooling1D(2),
    Flatten(),
    Dense(64, activation='relu'),
    Dense(1, activation='sigmoid')
])

# Compile the model
cnn_model.compile(optimizer=Adam(), loss='binary_crossentropy', metrics=['accuracy'])

# Train the model
history = cnn_model.fit(X_train, Y_train, epochs=10, batch_size=32, verbose=1, validation_split=0.2)

# Predict using the trained model
cnn_y_pred = (cnn_model.predict(X_test) > 0.5).astype(int).flatten()

# Calculate the accuracy
cnn_accuracy = accuracy_score(Y_test, cnn_y_pred)
print(f'CNN Model Accuracy: {cnn_accuracy:.2f}')
```

***ROC Curve (Blue Line)***:

The blue line represents the ROC curve, which plots the true positive rate (TPR) against the false positive rate (FPR) at various threshold settings.
The curve typically starts at the origin (0,0) and ends at the point (1,1).
The shape and position of the curve indicate the performance of the classifier. A curve that bows closer to the top-left corner indicates a better performance.

***Diagonal Line (Gray Dashed Line)***:

The gray dashed line represents a random classifier. It follows the line y=x, where the true positive rate equals the false positive rate.
This line serves as a baseline for evaluating the ROC curve. Any classifier performing above this line is better than random chance.

***AUC (Area Under the Curve)***:

The legend includes the AUC value, which is a single scalar value summarizing the performance of the classifier.
The AUC ranges from 0 to 1, with 1 indicating a perfect classifier and 0.5 indicating a classifier no better than random chance.
In this plot, the AUC is indicated in the legend (e.g., ROC curve (area = 0.90)). A higher AUC value indicates better classifier performance.
